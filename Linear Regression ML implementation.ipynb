```python
## packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Lets load boston housing price dataset

```python
boston_data=pd.read_csv("HousingData.csv")
```

```python
boston_data
```

```python
# Rename the column
boston_data.rename(columns={'MEDV': 'Price'}, inplace=True)
```

```python
boston_data
```

```python
boston_data.info()
```

```python
#### summarizing the stats 
boston_data.describe()
```

```python
### Check the missing values 
boston_data.isnull().sum()
```

```python
# Fill missing values with the mean of each column
df_filled_with_mean = boston_data.fillna(boston_data.mean())
```

```python
### Check the missing values 
df_filled_with_mean.isnull().sum()
```

```python
##### Find coorelation among dataset 
df_filled_with_mean.corr()
```

```python
#### Plot to see the corelation 
import seaborn as sns
sns.pairplot(df_filled_with_mean)
```

```python
##### analyzing individual factors 
plt.scatter(df_filled_with_mean['CRIM'],df_filled_with_mean['Price'])
plt.xlabel('Crime')
plt.ylabel('Price')
```

```python
plt.scatter(df_filled_with_mean['RM'],df_filled_with_mean['Price'])
plt.xlabel('Room')
plt.ylabel('Price')
```

```python
sns.regplot(x='LSTAT',y='Price',data=df_filled_with_mean)
```

```python
sns.regplot(x='CHAS',y='Price',data=df_filled_with_mean)
```

```python
sns.regplot(x='NOX',y='Price',data=df_filled_with_mean)
```

```python
##### Take out all independent and dependent features 

x=df_filled_with_mean.iloc[:,:-1]
y=df_filled_with_mean.iloc[:,-1]
```

```python
## devide data into train and test split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
### Standarize the dataset 
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
import pickle
pickle.dump(scaler,open('scaling.pkl','wb'))
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
##### Print the coefficients 

print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## n which params the model has been trained 

print(regression.get_params())
```

```python
#### Prediction on test data 

regression_predict=regression.predict(X_test)
```

```python
regression_predict
```

```python
#### Plot a scatter plot for predictions

plt.scatter(regression_predict,y_test)
```

```python
#### Residuals/ errors 
residuals=y_test-regression_predict
```

```python
#### Plot the residuals 
sns.displot(residuals,kind="kde")
```

```python
#### scater plot between predictions and residuals 
#### uniform distribution
plt.scatter(residuals,regression_predict)
```

```python
#### evaluation matrics

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

```python
print(mean_absolute_error(y_test,regression_predict))
print(mean_squared_error(y_test,regression_predict))

### print rmse 

print(np.sqrt(mean_squared_error(y_test,regression_predict)))
```

```python
#### R-square , and adj-R square

from sklearn.metrics import r2_score
score=r2_score(y_test,regression_predict)
print(score)
```

```python
#### adj-R square 

1-(1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
